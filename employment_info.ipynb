```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.laalmanac.com/employment/em03.php'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
table = soup.find('table')
```

```python
column_data = table.find_all('tr')
```

```python
list = []
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    list.append(individual_row_data)

list
```

```python
list
filtered_list = [item for item in list if item]
```

```python
filtered_list
```

```python
df = pd.DataFrame(filtered_list, columns=['City', 'Labor Force', 'Employed', 'Unemployed', 'Unemployed Rate (%)'])
```

```python
df
```

```python
df['City'] = df['City'].str.replace('†', '', regex=False)
```

```python
df
```

```python
def clean_and_convert(value):
    value = value.replace(',', '').rstrip('%').strip()  # Remove commas, trailing '%', and whitespace
    return pd.to_numeric(value, errors='coerce')  # Convert to numeric, setting errors to NaN
df.loc[:, df.columns != 'City'] = df.loc[:, df.columns != 'City'].applymap(clean_and_convert)
```

```python
df.to_csv('employment.csv', index=False)
```
