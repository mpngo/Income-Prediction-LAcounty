```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.laalmanac.com/population/po30.php'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
table = soup.find('table')
```

```python
table
```

```python
column_data = table.find_all('tr')
column_data
```

```python
list = []
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    list.append(individual_row_data)
    
```

```python
list
```

```python

filtered_list = [item for item in list if item]
```

```python
filtered_list
```

```python
df = pd.DataFrame(filtered_list, columns=['City', 'Total Households', 'Average Size', 'With Children Under 18', 'Married','Co-habiting Couple', 'Female only', 'Male Only'])
```

```python
df
```

```python
#df.to_csv('household_info.csv', index=False)
```

```python
def clean_and_convert(value):
    value = value.replace(',', '').rstrip('%').strip()  # Remove commas, trailing '+', and whitespace
    return pd.to_numeric(value, errors='coerce')  # Convert to numeric, setting errors to NaN
df.loc[:, df.columns != 'City'] = df.loc[:, df.columns != 'City'].applymap(clean_and_convert)
```
