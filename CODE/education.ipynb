```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.laalmanac.com/education/ed33aa.php'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
table = soup.find('table')
```

```python
column_data = table.find_all('tr')
```

```python
list = []
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    list.append(individual_row_data)
```

```python
filtered_list = [item for item in list if item]
```

```python
odd_index_elements = filtered_list[1::2]
```

```python
even_index_elements = filtered_list[::2]
```

```python
for sublist in odd_index_elements:
    sublist.pop(0)
    sublist.pop(0)
```

```python
for sublist in even_index_elements:
    sublist.pop(1)
```

```python
city_names = [sublist[0] for sublist in even_index_elements if sublist]
```

```python
df1 = pd.DataFrame(even_index_elements, columns=['City', 'Grade < 9', 'Grade 9-12','Highschool Diploma', 'Some College', 'Associate Degree', 'Undergraduate Degree', 'Graduate Degree'])
```

```python
df2 = pd.DataFrame(odd_index_elements, columns=['Grade < 9 (%)', 'Grade 9-12 (%)','Highschool Diploma (%)', 'Some College (%)', 'Associate Degree (%)', 'Undergraduate Degree (%)', 'Graduate Degree (%)'])
```

```python
df2.insert(0, 'City', city_names)
```

```python
df2
```

```python
merged_df = pd.merge(df1, df2, on='City', how='inner')
```

```python
merged_df
```

```python
def clean_and_convert(value):
    value = value.replace(',', '').rstrip('%').strip()  # Remove commas, trailing '+', and whitespace
    return pd.to_numeric(value, errors='coerce')  # Convert to numeric, setting errors to NaN

merged_df.loc[:, merged_df.columns != 'City'] = merged_df.loc[:, merged_df.columns != 'City'].applymap(clean_and_convert)
```

```python
merged_df['City'] = merged_df['City'].str.replace('*', '', regex=False)
```

```python
merged_df
```

```python
merged_df.to_csv('education.csv', index=False)
```
