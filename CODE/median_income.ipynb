```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.laalmanac.com/employment/em12.php'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
table = soup.find('table')
```

```python
table
```

```python
column_data = table.find_all('tr')
column_data
```

```python
list = []
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    list.append(individual_row_data)
    
```

```python
df = pd.DataFrame(list, columns=['City', 'All', 'Native', 'Asian','African Amr.', 'Pacific', 'White', 'Other', 'Two or More', 'Hispanic'])
```

```python
df
```

```python
df['City'] = df['City'].str.replace('*', '', regex=False)
df['All'] = df['All'].str.replace('$', '', regex=False)
df['Native'] = df['Native'].str.replace('$', '', regex=False)
df['Asian'] = df['Asian'].str.replace('$', '', regex=False)
df['African Amr.'] = df['African Amr.'].str.replace('$', '', regex=False)
df['Pacific'] = df['Pacific'].str.replace('$', '', regex=False)
df['White'] = df['White'].str.replace('$', '', regex=False)
df['Other'] = df['Other'].str.replace('$', '', regex=False)
df['Two or More'] = df['Two or More'].str.replace('$', '', regex=False)
df['Hispanic'] = df['Hispanic'].str.replace('$', '', regex=False)

df= df.dropna(subset=['City'])


df
```

```python
subset_df = df[['City', 'All']]
```
