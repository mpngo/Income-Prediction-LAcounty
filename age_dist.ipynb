```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.laalmanac.com/population/po09a.php'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
soup
```

```python
table = soup.find('table')
table
```

```python
column_data = table.find_all('tr')
column_data
```

```python
list = []
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    list.append(individual_row_data)
```

```python
list
```

```python
list
filtered_list = [item for item in list if item]
```

```python
filtered_list
```

```python
odd_index_elements = filtered_list[1::2]
odd_index_elements
```

```python
for sublist in odd_index_elements:
    sublist.pop(0)
```

```python
odd_index_elements
```

```python
df = pd.DataFrame(odd_index_elements, columns=['0-15 (%)', '15-17 (%)','18-24 (%)', '25-34 (%)', '35-54 (%)', '55-64 (%)', '65+ (%)'])
```

```python
df
```

```python
even_index_elements = filtered_list[::2]
even_index_elements
```

```python
city_names = [sublist[0] for sublist in even_index_elements if sublist]
```

```python
city_names
```

```python
df.insert(0, 'City', city_names)

#df
```

```python
df['City'] = df['City'].str.replace('â ', '', regex=False)
```

```python
new_data = pd.DataFrame(even_index_elements, columns=['City', 'Total Pop', '0 to 15', '15 to 17','18 to 24', '25 to 34', '35 to 54', '55 to 64', '65 plus'])
```

```python
new_data['City'] = new_data['City'].str.replace('â ', '', regex=False)
```

```python
merged_df = pd.merge(df, new_data, on='City', how='inner')
```

```python
def clean_and_convert(value):
    value = value.replace(',', '').rstrip('%').strip()  # Remove commas, trailing '+', and whitespace
    return pd.to_numeric(value, errors='coerce')  # Convert to numeric, setting errors to NaN

merged_df.loc[:, merged_df.columns != 'City'] = merged_df.loc[:, merged_df.columns != 'City'].applymap(clean_and_convert)
```

```python
merged_df
```

```python
merged_df.to_csv('age_dist.csv', index=False)
```
